# My Screen Time Dashboard Analysis

```
The data contains the following fields:

```
1. date : Date
2. week_number: Number of Week
3. facebook: Hours spent on Facebook
4. whatsapp: Hours spent on WhatsApp
5. youtube: Hours spent on Youtube
6. linkedin: Hours spent on Linkedin
7. gmail: Hours spent on Linkedin
8. adobe_acrobat: Hours spent on Adobe Acrobat
9. binance: Hours spent on Binance
10. samsung_internet: Hours spent on Samsung Internet
11. twitter: Hours spent on Twitter
12. notifications: Total notifications per day
13. screen_time: Total Screen Time per day
14. most_notifications_app: Most notification by an app per day

**TechStack:** 
* Numpy
*  Pandas
*  Matplotlib
*  Seaborn
